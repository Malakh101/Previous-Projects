```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('bestsellers with categories.csv')
data
```

```python
data.info()
```

```python
data.columns
```

```python
data['Year'] = pd.to_datetime(data['Year'], format='%Y')
```

```python
data.info()
```

```python
books_by_year=data.sort_values("Year")
books_by_year
```

```python
books_by_year.describe()
```

```python
best_selling=books_by_year.sort_values('Reviews',ascending=False).head(20)
best_selling
```

```python
best_selling.describe()
```

```python
plt.title('Most Reviewed')
sns.barplot(y=best_selling.Name, x=best_selling.Reviews)

```
